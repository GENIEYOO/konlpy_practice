```python
from konlpy.corpus import kobill
from konlpy.tag import Twitter; t = Twitter()
from matplotlib import pyplot as plt
from pprint import pprint
```

```python
text = "질량의 크기는 부피와 비례하지 않는다. 제비꽃같이 조그마한 그 계집애가 꽃잎같이 하늘거리는 그 계집애가지구보다 더 큰 질량으로 나를 끌어당긴다.순간, 나는뉴턴의 사과처럼사정없이 그녀에게로 굴러 떨어졌다쿵 소리를 내며, 쿵쿵 소리를 내며심장이하늘에서 땅까지 아찔한 진자운동을 계속하였다첫사랑이었다."
a = set(t.pos(text))
b = t.pos(text)
# pprint (a)
print (len(a))
print (len(b))
```

말 뭉치의 크기를 늘려갈수록 등장하는 토큰의 개수가 로그함수적으로 늘어간다는 힙의 법칙

```python
pos = lambda x: ['/'.join(p) for p in t.pos(x)]
#트위터 태그로 x를 태깅한 것을 하나씩 꺼내줘서 
docs = [kobill.open(i).read() for i in kobill.fileids()]
#kobill에 있는 파일명을 하나씩 찾아서 i에 넣고 각각의 파일을 하나씩 열어서 읽어라 그걸 리스트안에 넣어라
# print (docs)
```

첫 번째 파일 열어보기

```python
kobill.open(kobill.fileids()[0]).read()
```

open할 때 kobill에서 열어야함. 안그러면 디렉토리 오류남

```python
global_unique=[]
global_unique_cnt=[]
for doc in docs:
    tokens=pos(doc) #단어 + 태킹
    unique = set(tokens)#태깅한 것 중에서 중복단어 제거
    global_unique += list(unique) #unigue를 리스트화
    global_unique = list(set(global_unique)) #중복제거 unique하게 태깅 매겨진 것
    global_unique_cnt.append(len(global_unique))#unique태깅들 갯 수 세기
    print(len(unique), len(global_unique), global_unique)
    
plt.plot(global_unique_cnt)
plt.savefig('heap.png')

    
```
